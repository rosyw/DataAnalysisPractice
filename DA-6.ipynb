```python
import pandas as pd
import numpy as np
task1 = pd.read_csv('./task6_1.csv')
task1
```

```python
#练习
pd.pivot_table(task1,index=['id','date'])  #同一日的sale显示平均值
pd.pivot_table(task1,index=['id','goods']) #每类商品显示第一个日期sales
pd.pivot_table(task1,index=['id','date','goods']) 
pd.pivot_table(task1,index=['id','date','goods'],values=['sales']) #跟上面相同,values默认为sales？
```

```python
#每个人每天销售商品 A 和 B 的金额之和
task1.pivot_table(index=['id','date'],values=['sales'],aggfunc=[np.sum]) #每个人每天的销售总额
```

```python
#每个人的销售总金额
task1.pivot_table(index=['id'],values=['sales'],columns=['date','goods'],aggfunc=[np.sum],fill_value=0,margins=True,margins_name='all_sales') 
```

```python
task2 = pd.read_csv('./task6_2.csv')
task2
```

```python
#以性别为分组依据，查看男女抽烟平均年龄和身高
task2.pivot_table(index=['sex','smoke'],values=['age','height'],aggfunc=[np.mean])
```

```python
#利用交叉表知识，统计各个性别抽烟的人数（sex和smoke是参数）
pd.crosstab(task2.sex,task2.smoke)
```

```python
#利用交叉表知识，统计各个年龄段抽烟人情况(age和smoke是参数)
pd.crosstab(task2.age,task2.smoke)
```
