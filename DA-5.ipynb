```python
import pandas as pd
#数组列表创建df
data1 ={
    'marketcap': [449, 371, 237, 21313, 1369, 823], 
    'pe': [8.31, 15.36, 16.01, 7.16, 7.59, 6.3],
    'code': ['600926', '002958', '601128', '601398', '601229', '600919']
}
index = ['杭州银行', '青农商行', '常熟银行', '工商银行', '上海银行', '江苏银行']
df_company = pd.DataFrame(data1,index = index)
```

```python
df_company
```

```python
#marketcap不加引号
df_company.loc[df_company.marketcap<2000]
```

```python
#每个条件加括号，用&而不是and
df_company.loc[(df_company.marketcap<2000) & (df_company.pe<10)]
```

```python
import numpy as np

dr = pd.date_range(start='2019-01-02', periods=100)

data = np.random.randn(100).cumsum()

close = data - np.min(data)

df = pd.DataFrame({"close": close}, index=dr)

df
```

```python
df['roll_mean'] = df['close'].rolling(window=5).mean()
df
```

```python
#任务二
#二维数组直接创建
data2 = np.arange(1,17).reshape(4,4)
df_pra = pd.DataFrame(data2, index = list('abcd'), columns = ['第1列','第2列','第3列','第4列'])
df_pra
```

```python
#点多次总和会累加
df_pra['总和'] = df_pra.apply(lambda x: x.sum(),axis=1)

df_pra
```

```python
#位置索引是从0开始
#插入的参数df.insert(loc,column_name,values)
df_pra.insert(4,'第5列',df_pra.apply(lambda x: x[0]+x[1] if x[2]>10 else x[2],axis=1))
df_pra
```
