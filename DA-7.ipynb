```python
import pandas as pd
import numpy as np
raw = pd.read_csv('./jobs.csv')
new = raw.drop_duplicates(subset=['company','industry'],keep='last') #以公司为重复项，显示保留第一项
new
```

```python
new.head(10)
```

```python
#统计数据表中的招聘岗位都来自于哪些网站,normalise比例，sort分类项是否按a-z排序，ascending结果升降序
new['source'].value_counts()
```

```python
#数据分析师岗位工作经验的要求
new['experience'].value_counts()
```

```python
#对于教育背景的要求
new['education'].value_counts(sort=False)
```

```python
#不同类型公司的职位数量
new['company_type'].value_counts()
```

```python
#所有职位的平均工资是多少
raw['salary2'].mean()
```

```python
#统计平均薪资（即 salary2）排在前 10 位的招聘公司及金额
raw.groupby('company')['salary2'].mean().sort_values(ascending=False).head(10)
```

```python
#统计发布职位数排在前 10 位的招聘公司及其职位数量
raw['company'].value_counts(ascending=False).head(10)
```

```python
#统计出平均工资排在前 10 位的公司类型及平均工资金额
raw.groupby('company_type')['salary2'].mean().sort_values(ascending=False).head(10)
```
