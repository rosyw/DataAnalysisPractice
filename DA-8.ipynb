```python
import pandas as pd
raw = pd.read_csv('./houses.csv',encoding='gbk')
raw
```

```python
new = raw.drop_duplicates(['title','xiaoqu_name'],'last') #多个去重字段，去其重复的并集
new
```

```python
#2总房源数，直接用len()
len(new) 
```

```python
#小区数量
len(new['xiaoqu_name'].value_counts())
```

```python
#总板块数量
len(new['sub_district_name'].value_counts())
```

```python
#3不同户型，unique()
new['huxing'].unique()
```

```python
#户型数量nuique()
new['huxing'].nunique()
```

```python
#不同装修情况
new['zhuangxiu'].unique()
```

```python
#不同建造年份
new['buildyear'].unique()
```

```python
#每平米均价
new['up_price'].mean()
```

```python
#5不同户型对应的房源数量、平均面积
new.groupby('huxing')['size'].agg(['mean','count']).sort_values('count',ascending=False)
```

```python
#6获取每个小区的房源数量，并按从大到小的顺序排序,两种写法结果相同
new.groupby('xiaoqu_name')['title'].count().sort_values(ascending=False)
new['title'].groupby(new['xiaoqu_name']).count().sort_values(ascending=False)
```

```python
#7上海每个小区每平方米的均价
new.groupby('xiaoqu_name')['up_price'].mean()
```

```python
#8上海二手房总价排在前 5 位的房源，看看超级豪宅有多壕？
new.sort_values(by=['price'],ascending=False).head(5)
```

```python
#9不同年代的房子数量（10年一个区间）
#new.buildyear = new.buildyear.astype(float)
new.buildyear = pd.to_datetime(new.buildyear,format = '%Y')
new.buildyear  #dtype为datatime64
```

```python
year = new.groupby('buildyear')['title'].count()
year.resample('10AS').sum().fillna(0)  #已起始年算
```
